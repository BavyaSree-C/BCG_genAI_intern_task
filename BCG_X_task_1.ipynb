```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Company': ['Microsoft', 'Microsoft', 'Microsoft', 'Tesla', 'Tesla', 'Tesla', 'Apple', 'Apple', 'Apple'],
    'Total Revenue': [245122, 211915, 198270, 97690, 96773, 81462, 391035, 383285, 394328],
    'Net Income': [88136, 72361, 72738, 7153, 14974, 12587, 93736, 96995, 99803 ],
    'Total Assets': [512163, 411976, 364840, 122070, 106618, 82338, 364980, 352583, 352755 ],
    'Total Liabilities': [243686, 205753, 198298, 48390, 43009, 36440, 308030, 290437, 302083	],
    'Cash Flow': [118548, 87582, 89035, 14923, 13256, 14724, 30737, 24977, 35929],
    'Year': [2024, 2023, 2022, 2024, 2023, 2022, 2024, 2023, 2022]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df['Revenue Growth (%)'] = df.groupby(['Company'])['Total Revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net Income'].pct_change() * 100
```

```python
df.info()
```

```python
df = df.sort_values(by=['Company', 'Year'])
```

```python
df['Assets Growth (%)'] = df.groupby('Company')['Total Assets'].pct_change() * 100
df['Liabilities Growth (%)'] = df.groupby('Company')['Total Liabilities'].pct_change() * 100
df['Cash Flow Growth (%)'] = df.groupby('Company')['Cash Flow'].pct_change() * 100
```

```python
df
```

```python
df.fillna(0,inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.to_csv('final_BCG_report.csv')
```

```python
df
```

```python
# Aggregate statistics for each company
agg_df = df.groupby('Company').agg({
    'Total Revenue': ['mean', 'max', 'min'],
    'Net Income': ['mean', 'max', 'min'],
    'Revenue Growth (%)': ['mean'],
    'Net Income Growth (%)': ['mean'],
    'Assets Growth (%)' : ['mean'],
    'Liabilities Growth (%)' : ['mean'],
    'Cash Flow Growth (%)' :['mean']
}).reset_index()

print(agg_df)
```

```python
agg_df
```

```python
agg_df.to_csv('Summary_final_report.csv')
```
